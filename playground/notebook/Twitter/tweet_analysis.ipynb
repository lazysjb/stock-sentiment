```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_tsla_tweets_for_date(date_str):
    file_path = 'data/tweets_{}_$TSLA.json'.format(date_str)
    return pd.read_json(file_path, dtype=False)
```

```python
def get_tweet_count_for_date(date_str):
    df = read_tsla_tweets_for_date(date_str)
    return df.shape[0]
```

```python
start = '2018-01-01'
end = '2018-06-06'
```

```python
tweet_counts = dict()

for date in pd.date_range(start, end):
    date_str = date.strftime('%Y-%m-%d')
    tweet_counts[date_str] = get_tweet_count_for_date(date_str)
    
```

```python
count_df = pd.Series(tweet_counts, name='count').to_frame()
count_df.index = pd.to_datetime(count_df.index)
```

```python
count_df.plot()
```
