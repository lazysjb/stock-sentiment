```python
ts = TimeSeries(key='BZ06I447HB2IFDC7', output_format="pandas")
```

```python
stock_df, meta_data = ts.get_daily(ticker,outputsize="full")
```

```python
stock_df.index = pd.to_datetime(stock_df.index)
```

```python
def clean_column_names(col_name):
    components = col_name.split('. ')
    return components[-1]
```

```python
stock_df.columns = [clean_column_names(c) for c in stock_df.columns]
```

```python
stock_df.loc['2014': '2018', 'close'].plot()
```

```python
stock_df.loc['2018-05-14': , 'close'].plot()
```
